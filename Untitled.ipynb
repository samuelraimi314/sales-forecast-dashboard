```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

# --- 1. Load Data ---
print("Loading data...")
try:
    df_train = pd.read_csv('train.csv')
    df_stores = pd.read_csv('stores.csv')
    df_features = pd.read_csv('features.csv')
    print("Data loaded successfully.")
except FileNotFoundError:
    print("Error: Ensure 'train.csv', 'stores.csv', and 'features.csv' are in the current directory.")
    # Exit or use sample data if necessary
    # return

# Convert Date columns to datetime objects
df_train['Date'] = pd.to_datetime(df_train['Date'])
df_features['Date'] = pd.to_datetime(df_features['Date'])

# --- 2. Merge DataFrames ---
# Merge train and features on Store and Date (many-to-one merge)
df_merged = pd.merge(df_train, df_features, on=['Store', 'Date'], how='left')

# Merge the result with stores on Store (many-to-one merge)
df_merged = pd.merge(df_merged, df_stores, on='Store', how='left')

# Drop the 'IsHoliday_x' column from train.csv (we use the 'IsHoliday_y' from features.csv)
# The features.csv 'IsHoliday' is generally more complete/accurate for external factors.
df_merged = df_merged.drop(columns=['IsHoliday_x'])
df_merged = df_merged.rename(columns={'IsHoliday_y': 'IsHoliday'})

print(f"\nMerged DataFrame Shape: {df_merged.shape}")
print("Columns:", df_merged.columns.tolist())
```

```python

# --- 3. Time and Categorical Features ---

# Extract Time Features
df_merged['Year'] = df_merged['Date'].dt.year
df_merged['Month'] = df_merged['Date'].dt.month
df_merged['Week'] = df_merged['Date'].dt.isocalendar().week.astype(int)
df_merged['Day'] = df_merged['Date'].dt.day
# Week of Year is often better than Month for seasonality

# Categorical Feature Encoding
# Convert IsHoliday (boolean) to integer (0 or 1)
df_merged['IsHoliday'] = df_merged['IsHoliday'].astype(int)

# One-Hot Encode Store Type (Type A, B, C)
df_merged = pd.get_dummies(df_merged, columns=['Type'], prefix='Store_Type', drop_first=True)
# drop_first=True avoids multicollinearity (Type B and C being 0 means it's Type A)

# Combine Store and Dept into a single ID for convenience (optional, but good for grouping)
df_merged['Store_Dept'] = df_merged['Store'].astype(str) + '_' + df_merged['Dept'].astype(str)

print("\nEngineered Time and Categorical Features.")
print(df_merged[['Date', 'Year', 'Week', 'Store_Type_B', 'Store_Type_C', 'IsHoliday']].head())
```

```python
# --- 4. Lagged Features (Most Important for Time Series) ---

# Sort data is CRITICAL for time-series operations
df_merged = df_merged.sort_values(by=['Store', 'Dept', 'Date']).reset_index(drop=True)

# Define lags. Lag-1 is sales from the previous week.
LAG_WEEKS = [1, 2, 4, 12, 52] # Common lags: Last week, 2 weeks, 1 month, 3 months, 1 year

print(f"\nCreating Lagged Sales Features: {LAG_WEEKS} weeks ago...")

# Apply the shift/lag operation, grouped by the individual time series (Store_Dept)
for lag in LAG_WEEKS:
    # Use Store and Dept to group, ensuring we only lag sales within the same time series
    df_merged[f'Sales_Lag_{lag}W'] = df_merged.groupby(['Store', 'Dept'])['Weekly_Sales'].shift(lag)
    
# Fill NaNs created by lagging (the first few weeks of each series won't have a lag)
# We can fill with 0 or the mean, but for now, we'll keep NaNs as they will be handled by XGBoost
# A better approach is to drop these initial rows if they cover a very small fraction of data,
# but for now, we'll let the model handle them.

print("Lagged features creation complete.")
print(df_merged[['Date', 'Weekly_Sales', 'Sales_Lag_1W', 'Sales_Lag_52W']].tail())
```

```python
# --- 5. Final Cleanup and Missing Value Inspection ---

# Inspect the dataset for missing values (from lags and feature files)
print("\n--- Missing Value Report ---")
missing_counts = df_merged.isnull().sum()
missing_counts = missing_counts[missing_counts > 0].sort_values(ascending=False)
print(missing_counts)

# Note: The NaNs in Markdown and CPI/Unemployment come from 'features.csv'
# The NaNs in Sales_Lag_... come from the start of each time series.

# For simplicity in the initial model:
# a) Drop rows with NaN in the TARGET VARIABLE (Weekly_Sales) - although there shouldn't be any in train.csv
df_merged = df_merged.dropna(subset=['Weekly_Sales'])

# b) Fill NaNs in external regressors (Markdowns, CPI, etc.) with 0 or the median/mean.
# Markdowns are often 0 when no promotion is running.
df_merged['MarkDown1'] = df_merged['MarkDown1'].fillna(0)
df_merged['MarkDown2'] = df_merged['MarkDown2'].fillna(0)
df_merged['MarkDown3'] = df_merged['MarkDown3'].fillna(0)
df_merged['MarkDown4'] = df_merged['MarkDown4'].fillna(0)
df_merged['MarkDown5'] = df_merged['MarkDown5'].fillna(0)

# Other features (CPI, Unemployment, Temperature, Fuel_Price) should be filled with median/mean
# for the respective store/date/overall to avoid data leakage in a real-world setting.
# For simplicity here, let's use the overall median for external regressors:
for col in ['CPI', 'Unemployment', 'Temperature', 'Fuel_Price']:
    df_merged[col] = df_merged[col].fillna(df_merged[col].median())


# --- Save the Cleaned Feature Set ---
FINAL_FILE = 'walmart_engineered_features.csv'
df_merged.to_csv(FINAL_FILE, index=False)
print(f"\n--- SUCCESS ---")
print(f"Final Feature DataFrame saved to {FINAL_FILE}")
print(f"Final shape: {df_merged.shape}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, make_scorer

def weighted_mean_absolute_error(y_true, y_pred, weights):
    """
    Calculates the Weighted Mean Absolute Error (WMAE).
    
    WMAE = (1 / sum(weights)) * sum(weights * |y_true - y_pred|)
    
    Args:
        y_true (pd.Series or np.array): The true Weekly_Sales values.
        y_pred (pd.Series or np.array): The predicted Weekly_Sales values.
        weights (pd.Series or np.array): The weights (5 for holiday, 1 otherwise).
        
    Returns:
        float: The calculated WMAE.
    """
    # Ensure all inputs are NumPy arrays for calculation efficiency
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    weights = np.array(weights)
    
    # Calculate the weighted absolute difference
    weighted_abs_diff = weights * np.abs(y_true - y_pred)
    
    # Calculate WMAE
    wmae = np.sum(weighted_abs_diff) / np.sum(weights)
    return wmae

print("WMAE function defined successfully.")
```

```python
# Load the cleaned data
df = pd.read_csv('walmart_engineered_features.csv')

# --- Define Features (X) and Target (y) ---
# Drop non-feature columns and features that introduce leakage (Date, Store_Dept, etc.)
FEATURES = [
    # Core Identifiers
    'Store', 'Dept', 'Size', 
    # Time Features
    'Year', 'Month', 'Week', 'Day', 'IsHoliday',
    # External Economic/Weather Features
    'Temperature', 'Fuel_Price', 'CPI', 'Unemployment',
    # Promotion Features (Markdowns)
    'MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5',
    # Engineered Lag Features (CRITICAL)
    'Sales_Lag_1W', 'Sales_Lag_2W', 'Sales_Lag_4W', 'Sales_Lag_12W', 'Sales_Lag_52W',
    # Store Type Encoding
    'Store_Type_B', 'Store_Type_C'
]

TARGET = 'Weekly_Sales'
WEIGHT = 'IsHoliday' # We will use the IsHoliday column as the weight proxy (0 or 1), 
                     # but need to adjust the value to 5 for holidays in the WMAE calculation later.

# Filter out rows where lagged sales are NaN (cannot be predicted without history)
df_model = df.dropna(subset=['Sales_Lag_52W']).copy()

# Filter out negative sales (often store returns/issues, treat as noise or 0)
df_model.loc[df_model[TARGET] < 0, TARGET] = 0

X = df_model[FEATURES]
y = df_model[TARGET]
weights = df_model[WEIGHT]

# --- Time-Based Train-Validation Split ---
# Use the last few months (e.g., 6 months of data or the last 15-20 weeks) for validation
# The latest date in the dataset is 2012-10-26 (Week 43 of 2012)
# We will use data up to 2012-07-01 for training, and the rest for validation.

# Get the split index for roughly the last 15-20 weeks (adjust as needed)
SPLIT_DATE = '2012-07-01' 

X_train = df_model[df_model['Date'] < SPLIT_DATE][FEATURES]
y_train = df_model[df_model['Date'] < SPLIT_DATE][TARGET]
weights_train = df_model[df_model['Date'] < SPLIT_DATE]['IsHoliday'].apply(lambda x: 5 if x == 1 else 1)
weights_train_raw = df_model[df_model['Date'] < SPLIT_DATE]['IsHoliday'] # Raw 0/1 for XGBoost

X_val = df_model[df_model['Date'] >= SPLIT_DATE][FEATURES]
y_val = df_model[df_model['Date'] >= SPLIT_DATE][TARGET]
weights_val = df_model[df_model['Date'] >= SPLIT_DATE]['IsHoliday'].apply(lambda x: 5 if x == 1 else 1) # Weights for WMAE calculation

print(f"\n--- Data Split Summary ---")
print(f"Total Rows: {len(df_model)}")
print(f"Training Data (before {SPLIT_DATE}): {len(X_train)} rows")
print(f"Validation Data (after {SPLIT_DATE}): {len(X_val)} rows")

# Check for remaining NaNs (XGBoost can sometimes handle NaNs, but clean inputs are best)
# X_train = X_train.fillna(0) # Use this if simple imputation is needed
# X_val = X_val.fillna(0)
```

```python
print("\n--- Training XGBoost Model ---")

model = XGBRegressor(
    n_estimators=1000,
    max_depth=7,
    learning_rate=0.05,
    subsample=0.7,
    colsample_bytree=0.7,
    random_state=42,
    tree_method='hist',
    n_jobs=-1,
    eval_metric='mae'   # ✔️ MUST go here in XGBoost 3.x
)

model.fit(
    X_train,
    y_train,
    eval_set=[(X_val, y_val)],  # ✔️ allowed
    verbose=False
)

print("Training complete.")
```

```python
import xgboost as xgb

print("\n--- Training XGBoost Model (Native API) ---")

# Convert data to DMatrix
dtrain = xgb.DMatrix(X_train, label=y_train)
dval = xgb.DMatrix(X_val, label=y_val)

# Model parameters
params = {
    "objective": "reg:squarederror",   # regression objective
    "eval_metric": "mae",              # use MAE for monitoring
    "tree_method": "hist",             # fast histogram algorithm
    "learning_rate": 0.05,
    "max_depth": 7,
    "subsample": 0.7,
    "colsample_bytree": 0.7,
    "seed": 42
}

# Train with early stopping
model = xgb.train(
    params=params,
    dtrain=dtrain,
    num_boost_round=1000,              # max trees
    evals=[(dtrain, "train"), (dval, "val")],
    early_stopping_rounds=50,          # stops based on eval metric
    verbose_eval=False                 # silence logs
)

print("Training complete.")
print(f"Best iteration: {model.best_iteration}")

# ---- Predict on validation set ----
y_pred_val = model.predict(dval)

# Clip negative predictions (sales cannot be negative)
y_pred_val = np.maximum(y_pred_val, 0)

# ---- Compute WMAE ----
final_wmae = weighted_mean_absolute_error(y_val, y_pred_val, weights_val)

print("\n--- Validation Results ---")
print(f"Final WMAE: ${final_wmae:,.2f}")
print(f"Overall MAE: ${mean_absolute_error(y_val, y_pred_val):,.2f}")
print(f"WMAE % of Mean Sales: {(final_wmae / y_val.mean()) * 100:.2f}%")
print(f"Mean Weekly Sales: ${y_val.mean():,.2f}")
```
